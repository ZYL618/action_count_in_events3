```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
from pylab import mpl 
mpl.rcParams["font.sans-serif"] = ["SimHei"]
```

```python
def compute_distance(x1,y1,x2,y2):
    return abs(x1-x2),abs(y1-y2)
# 通过与下一个事件点比较从而判断当前事件点是否为噪声
def is_noise(row,next_event,delta_x,delta_y):
    # 当前点的下一个点距离太远且count==1，则当前点被认为是噪声点
    x_dis,y_dis = compute_distance(row[1],row[2],next_event[1],next_event[2])
    # 当前事件点是噪声，跳过不录入
    if x_dis > delta_x and y_dis > delta_y:
        # 是噪声点
        return True
    else:
        return False
def write_to_csv(baseline_timestamp,temp_x,temp_y,count,path):
    pd.DataFrame(data=[[baseline_timestamp,temp_x/count,temp_y/count]]).to_csv(path,
                                                   mode='a',header=False,index=False)
def select_start_point(df,delta_x,delta_y):
    for index,row in df.iterrows():
        d1,d2 = compute_distance(row[1],row[2],df['x'][index+1],df['y'][index+1])
        if d1 <= delta_x and d2 < delta_y:
            # 说明当前点周围有其他点，大概率不是噪声点
            return index,row        
```

```python
# 
def compress_by_Manhattan(file_path,file_name,to_file_path,delta_x,delta_y):
    df = pd.read_csv(file_path)
    # 曼哈顿距离比较对象【最开始的点可能是噪声点】
    start,row = select_start_point(df,delta_x,delta_y)
    baseline_timestamp = row[0]
#     baseline_timestamp = df['timestamp'][0]
    baseline_x = row[1]
#     baseline_x = df['x'][0]
    baseline_y = row[2]
#     baseline_y = df['y'][0]
    
    # 待写入的曼哈顿距离
    temp_x = 0
    temp_y = 0
    count = 0
    pd.DataFrame(data=[['timestamp','x','y']]).to_csv(to_file_path,
                                                   mode='w',header=False,index=False)
    # 遍历每一行
    for index,row in df.iterrows():
        if index < start:
            continue
        # 在曼哈顿距离内，聚为一个点，取平均值
        x_dis,y_dis = compute_distance(baseline_x,baseline_y,row[1],row[2])
        # 最多聚焦 10 个点
        if x_dis <= delta_x and y_dis <= delta_y and count < 100:
            # 平均值的分母
            count = count + 1
            temp_x = temp_x + row[1]
            temp_y = temp_y + row[2]
        else:
            if index + 1 >= df.shape[0]:
                break
            # 当前点的下一个点距离太远且count==1，则当前点被认为是噪声点
            next_event = df.iloc[index+1]
            # 当前事件点是噪声，跳过不录入
            if count == 1 and is_noise(row,next_event,delta_x,delta_y):
                continue
            write_to_csv(baseline_timestamp,temp_x,temp_y,count,to_file_path)
            baseline_timestamp = row[0]
            baseline_x = row[1]
            baseline_y = row[2]
            temp_x = row[1]
            temp_y = row[2]
            count = 1
```

```python
# 绘制对照组
def plot_x_y(orignal_x,x,orignal_y,y,pre_title=''):
    plt.figure(figsize=(12,8),dpi=100)
    plt.subplot(221)
    plt.plot(range(len(orignal_x)),orignal_x,color='g',label='orignal_x',lw=0.1)
    plt.legend(loc=0)
    plt.title(f'{pre_title}初始x波形')
    
    plt.subplot(222)
    plt.plot(range(len(x)),x,color='r',label='x',lw=0.1)
    plt.legend(loc=0)
    plt.title(f'{pre_title}Δ=5，count=100压缩后x波形')
    
    plt.subplot(223)
    plt.plot(range(len(orignal_y)),orignal_y,color='#FF6A6A',label='orignal_y',lw=0.1)
    plt.legend(loc=0)
    plt.title(f'{pre_title}初始y波形')
    
    plt.subplot(224)
    plt.plot(range(len(y)),y,label='y',lw=0.1)
    plt.legend(loc=0)
    plt.title(f'{pre_title}Δ=5，count=100压缩后y波形')
    
    plt.show()
# 无对照组
def plot_x_y2(x,y):
    plt.figure(figsize=(12,8),dpi=100)
    plt.subplot(211)
    plt.plot(range(len(x)),x,color='g',label='orignal_x',lw=0.1)
    plt.legend(loc=0)
    plt.title('初始x波形')
    
    plt.subplot(212)
    plt.plot(range(len(y)),y,color='#FF6A6A',label='orignal_y',lw=0.1)
    plt.legend(loc=0)
    plt.title('初始y波形')
    
    plt.show()
```

```python
# 重复次数
repeat_times = 5
data = pd.read_csv('../split_data/class1/user01_fluorescent.csv')
data = data.query('timestamp > 4e5 & timestamp < 8e5')
data1 = data
for i in range(repeat_times-1):
    data1 = pd.concat([data1, data],ignore_index=True)
data1.to_csv('user01f_0.4_0.8s.csv',index=False)
compress_by_Manhattan('user01f_0.4_0.8s.csv','user01f_0.4_0.8s.csv','compress_user01_fluorescent.csv',5,5)
```
