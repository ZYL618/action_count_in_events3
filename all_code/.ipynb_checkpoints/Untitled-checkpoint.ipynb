```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
y1 = pd.read_csv('../event_csv/compress_event_manhattan/classQUVA/fitness.csv')['y']
# HP滤波
_, y2 = sm.tsa.filters.hpfilter(y1)
# 均值压缩
y2 = pd.DataFrame(y2)
y2['group'] = y2.index // 200  
y3 = y2.groupby('group')['y_trend'].mean()
# 二次HP滤波
_, y4 = sm.tsa.filters.hpfilter(y3)
```

```python
plt.figure(figsize=(30,12),dpi=200)
plt.subplot(4,1,1)
plt.title('距离压缩后的波形')
plt.plot(range(y1.shape[0]),y1)
plt.subplot(4,1,2)
plt.title('第1次HP滤波后的波形')
plt.plot(range(y2.shape[0]),y2)
plt.subplot(4,1,3)
plt.title('均值处理后的波形')
plt.plot(range(y3.shape[0]),y3)
plt.subplot(4,1,4)
plt.title('第2次HP滤波后的波形')
plt.plot(range(y3.shape[0]),y3)
plt.plot(range(smooth_x.shape[0]),smooth_x)
plt.show()
```
